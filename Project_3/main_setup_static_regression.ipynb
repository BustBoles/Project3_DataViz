```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pathlib as path
```

```python
#Import CSV
aggregate = pd.read_csv('Resources/PROV_US_AGGREGATE20.CSV')
med_data = pd.DataFrame(aggregate)
med_data.head()
```

```python
#Rename Region Number, Region Name, and State columns
med_data.rename(columns = {'Region_Number': 'Region(#)', 'REGION_NAME': 'Region', 'State_Name': 'State'}, inplace = True)
med_data.head()
```

```python
#Remove 'Y' from all year columns
med_data.rename(columns = {'Y1980': '1980', 'Y1981': '1981', 'Y1982': '1982', 'Y1983': '1983',\
                          'Y1984': '1984', 'Y1985': '1985', 'Y1986': '1986', \
                          'Y1988': '1988', 'Y1989': '1989', 'Y1990': '1990', \
                          'Y1991': '1991', 'Y1992': '1992', 'Y1993': '1993', \
                          'Y1994': '1994', 'Y1995': '1995', 'Y1996': '1996', \
                          'Y1997': '1997', 'Y1998': '1998', 'Y1999': '1999', \
                          'Y2000': '2000', 'Y2001': '2001', 'Y2002': '2002', \
                          'Y2003': '2003', 'Y2004': '2004', 'Y2005': '2005', \
                          'Y2006': '2006', 'Y2007': '2007', 'Y2008': '2008', \
                          'Y2009': '2009', 'Y2010': '2010', 'Y2011': '2011', \
                          'Y2012': '2012', 'Y2013': '2013', 'Y2014': '2014', \
                          'Y2015': '2015', 'Y2016': '2016', 'Y2017': '2017', \
                          'Y2018': '2018', 'Y2019': '2019', 'Y2020': '2020', \
                          'Y2021': '2021', 'Y2022': '2022', 'Y1987': '1987'
                         
                        }, inplace = True)
med_data.head()
```

```python
#Rename Average Anual Growth column
med_data.rename(columns = {'Average_Annual_Percent_Growth': 'Avg An Growth(%)'}, inplace = True)
med_data
```

```python
# Drop columns we don't need
list = ['Code', 'Region(#)', 'Region']
med_data = med_data.drop(list, axis = "columns")
med_data.head()
```

```python
# Filter 'Group' column to just show state
med_data.dropna(subset=['State'], inplace = True)
med_data
```

```python
# Export to csv
med_data.to_csv("Resources/med_data.csv", index=False)
```

```python
# Print all entries
query = {'Group': 'State'}
med_data_all = med_data.find(query)
med_data_all_df = pd.DataFrame(med_data_all)
med_data_all_df.head()
```

```python
# Print column names
print("Column Names:")
print(med_data_all_df.columns)
```

```python
#New order of columns
new_order = ['_id', 'State', 'Group', 'Item', '1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987',
       '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996',
       '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005',
       '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014',
       '2015', '2016', '2017', '2018', '2019', '2020']

# Create a new DataFrame with columns in the desired order
med_data_all_df_reorder = med_data_all_df[new_order]
med_data_all_df_reorder.head()
```

```python
# Drop columns we don't need
list = ['_id', 'Group']
med_data_all_df_reorder = med_data_all_df_reorder.drop(list, axis = "columns")
med_data_all_df_reorder.head()
```

```python
# Reshape the DataFrame into long format
med_data_long_df = pd.melt(med_data_all_df_reorder, id_vars=['State', 'Item'], var_name='Year', value_name='Spending')
med_data_long_df.head()
```

```python
#Print DF
print(med_data_long_df)
```

```python
#Print to show all unqiue items in med_data_long_df
unique_items = med_data_long_df["Item"].unique()

for item in unique_items:
    print(item)
```

```python
#Import matplotlib
import matplotlib.pyplot as plt

# Filter the DataFrame for California
california_data = med_data_long_df[med_data_long_df['State'] == 'California']

# Establish 'Year' as type int
california_data['Year'] = california_data['Year'].astype(int)

# Setting the figure size
plt.figure(figsize=(12, 6))

# Define unique categories (Items)
categories = california_data['Item'].unique()

# Create for loop to define category_data variable
for category in categories:
    category_data = california_data[california_data['Item'] == category]
    category_data = category_data.sort_values('Year')
    plt.plot(category_data['Year'], category_data['Spending'], label=category, marker='o')

# Create title and labels
plt.title('Healthcare Spending by Category Over Time in California', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Spending (in Millions)', fontsize=14)

# Create legend
plt.legend()

# Plot to show
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter DF on 'Personal Health Care (Millions of Dollars)'
personal_healthcare_data = med_data_long_df[med_data_long_df['Item'] == 'Personal Health Care (Millions of Dollars)']


personal_healthcare_data['Year'] = personal_healthcare_data['Year'].astype(int)
personal_healthcare_data['Spending'] = pd.to_numeric(personal_healthcare_data['Spending'], errors='coerce')

# Sum of spending for each state for top 10 largest spenders
total_spending_by_state = personal_healthcare_data.groupby('State')['Spending'].sum().nlargest(10)
top_states = total_spending_by_state.index.tolist()

# Plot fig size
plt.figure(figsize=(15, 8))

# Create for loop to define category data
for state in top_states:
    state_data = personal_healthcare_data[personal_healthcare_data['State'] == state]
    state_data = state_data.sort_values('Year')
    plt.plot(state_data['Year'], state_data['Spending'], label=state, linewidth=2, marker='o')  # Adjusted line width and added markers

# Create title and labels
plt.title('Top 10 States in Personal Healthcare Spending Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Spending (in Millions)', fontsize=14)

# Create legend
plt.legend()

# Plot to show
plt.show()
```

```python
import pandas as pd
from bokeh.plotting import show
from bokeh.models import ColumnDataSource, DataTable, TableColumn
```

```python
#Create df variable to represent med_data_long_df filtered on state by name
df = med_data_long_df[med_data_long_df['State'] == 'California']

#Run Regression Panda
stats = df.describe().reset_index()
print(stats.reset_index())
```

```python
#Create source and column varaibles for DataTable
source = ColumnDataSource(stats)
columns = [TableColumn(field=col, title=col) for col in stats.columns]

#Print to show Bokeh DataTable
data_table = DataTable(source=source, columns=columns, width=800, height=400, index_position=None)
show(data_table)
```
